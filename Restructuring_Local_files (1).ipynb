```python
# Using OS and shutil to restructre folder hirearchy in local files
# debridge file path in flattened zip


import os
import shutil
from google.colab import files

def upload_and_flatten_folder(destination_folder='flattened_files'):
    print("Upload your zipped folder (multiple files/folders inside):")
    uploaded = files.upload()
    if not uploaded:
        print("No files uploaded.")
        return

    # There should be one uploaded zipped file (a folder zipped)
    zip_filename = list(uploaded.keys())[0]
    print(f"Uploaded zip file: {zip_filename}")

    # Unzip into a temporary folder
    import zipfile
    temp_extract_folder = 'temp_extracted_folder'
    os.makedirs(temp_extract_folder, exist_ok=True)

    with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
        zip_ref.extractall(temp_extract_folder)
    print(f"Extracted '{zip_filename}' to temporary folder '{temp_extract_folder}'")

    # Create destination folder for flattened files
    os.makedirs(destination_folder, exist_ok=True)

    # Walk through all files in extracted folder and copy to destination folder
    for root, dirs, files_in_dir in os.walk(temp_extract_folder):
        for file in files_in_dir:
            source_path = os.path.join(root, file)
            dest_path = os.path.join(destination_folder, file)

            # Avoid overwrite by renaming duplicates
            if os.path.exists(dest_path):
                base, ext = os.path.splitext(file)
                count = 1
                while os.path.exists(dest_path):
                    new_file = f"{base}_{count}{ext}"
                    dest_path = os.path.join(destination_folder, new_file)
                    count += 1

            shutil.copy2(source_path, dest_path)
            print(f"Copied {source_path} to {dest_path}")

    print(f"\nAll files flattened and saved into folder: '{destination_folder}'")

# Call the function
upload_and_flatten_folder()
# importing from cloud

import shutil
shutil.make_archive('flattened_files_zip', 'zip', 'flattened_files')

from google.colab import files
files.download('flattened_files_zip.zip')
```
